```python
import json
from pathlib import Path
```

```python
BASE_DIR = Path().cwd().parents[2]
PROCESSED_DATA_DIR = BASE_DIR / "data" / "processed_data" / "bedrooms" / "sklad_mebliv"
```

```python
scenes_processed = 0

for scene_folder in sorted(PROCESSED_DATA_DIR.glob("sklad_mebliv_[0-9][0-9][0-9][0-9]")):
    if not scene_folder.is_dir():
        continue

    annotation_path = scene_folder / "annotations.json"
    
    if annotation_path.exists():
        with open(annotation_path, "r", encoding="utf-8") as f:
            annotation = json.load(f)
    else:
        annotation = {
            "scene_name": scene_folder.name,
            "furnitures": []
        }
    
    # Convert all categories to lowercase
    for furniture in annotation.get("furnitures", []):
        furniture_href = furniture.pop("href", "")
        if furniture_href:
            furniture["furniture_href"] = furniture_href

        if "category" in furniture:
            cat = furniture["category"].lower()
            if cat == "chair#stool":
                cat = "chair_stool"
            furniture["category"] = cat

            
            if cat == 'nightstand':
                furniture["category"] = "small_storage"
            
            if cat == 'storage':
                furniture["category"] = "large_storage"
    
    with open(annotation_path, "w", encoding="utf-8") as f:
        json.dump(annotation, f, ensure_ascii=False, indent=2)
    
    scenes_processed += 1
    print(f"{scene_folder.name}")

print(f"Total scenes processed: {scenes_processed}")
```
