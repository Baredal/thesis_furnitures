```python
import json
from pathlib import Path
```

```python
BASE_DIR = Path().cwd().parents[2]  # Navigate to project root
PROCESSED_DATA_DIR = BASE_DIR / "data" / "processed_data" / "deepfurn"
```

```python
scenes_processed = 0

for scene_folder in sorted(PROCESSED_DATA_DIR.glob("deepfurn_[0-9][0-9][0-9][0-9]")):
    if not scene_folder.is_dir():
        continue

    # Read existing annotation.json if it exists
    annotation_path = scene_folder / "clip_annotations.json"

    if annotation_path.exists():
        with open(annotation_path, "r", encoding="utf-8") as f:
            annotation = json.load(f)
    else:
        # Create new annotation structure if missing
        annotation = {
            "scene_name": scene_folder.name,
            "furnitures": []
        }

    for furniture in annotation.get("furnitures", []):
        if "category" in furniture:
            cat = furniture["category"].lower()
            if cat == "chair#stool":
                cat = "chair_stool"
            furniture["category"] = cat

    # Write back the updated annotation
    with open(scene_folder / "annotations.json", "w", encoding="utf-8") as f:
        json.dump(annotation, f, ensure_ascii=False, indent=2)
    
    scenes_processed += 1
    print(f"{scene_folder.name}")

print(f"Total scenes processed: {scenes_processed}")
```
