```python
import os
import json
import shutil
from pathlib import Path
from tqdm import tqdm
```

Bedrooms

```python
input_dir = Path("bedroom_scenes")
output_dir = Path("initial_process_bedrooms_scenes")

# Categories to remove based on IDs 5, 8, 9, 11
exclude_categories = {"Door", "Plant", "Decoration", "Home-appliance", "home-appliance", "lamp", "Lamp"}

if not output_dir.exists():
    output_dir.mkdir(parents=True)

for scene_path in tqdm(list(input_dir.iterdir())):
    if not scene_path.is_dir():
        continue
        
    scene_id = scene_path.name
    scene_output_dir = output_dir / scene_id
    scene_output_dir.mkdir(parents=True, exist_ok=True)
    
    scene_img_in = scene_path / "scene_image.jpg"
    scene_img_out = scene_output_dir / "scene_image.jpg"
    if scene_img_in.exists() and not scene_img_out.exists():
        shutil.copy(scene_img_in, scene_img_out)
        
    # Process annotation.json
    ann_file_in = scene_path / "annotation.json"
    ann_file_out = scene_output_dir / "annotation.json"
    
    if ann_file_in.exists():
        with open(ann_file_in, "r", encoding="utf-8") as f:
            data = json.load(f)
            
        # Filter categories
        if "categories" in data:
            updated_categories = []
            for c in data["categories"]:
                if c.title() not in [x.title() for x in exclude_categories]:
                    cat_lower = c.lower()
                    if cat_lower == "chair#stool":
                        cat_lower = "chair_stool"
                    elif cat_lower == "cabinet#shelf":
                        cat_lower = "cabinet_shelf"
                    if cat_lower not in updated_categories:
                        updated_categories.append(cat_lower)
            data["categories"] = updated_categories
            
        if "furnitures" in data:
            valid_furnitures = []
            for item in data["furnitures"]:
                cat = item.get("category", "")
                if cat.title() not in [x.title() for x in exclude_categories]:
                    cat_lower = cat.lower()
                    if cat_lower == "chair#stool":
                        cat_lower = "chair_stool"
                    elif cat_lower == "cabinet#shelf":
                        cat_lower = "cabinet_shelf"
                    item["category"] = cat_lower
                    valid_furnitures.append(item)
                    
                    img_filename = item.get("image_filename")
                    if img_filename:
                        furn_img_in = scene_path / img_filename
                        furn_img_out = scene_output_dir / img_filename
                        
                        furn_img_out.parent.mkdir(parents=True, exist_ok=True)
                        if furn_img_in.exists() and not furn_img_out.exists():
                            shutil.copy(furn_img_in, furn_img_out)
                            
            data["furnitures"] = valid_furnitures
            
        if "boxes" in data:
            pass 
            
        with open(ann_file_out, "w", encoding="utf-8") as f:
            json.dump(data, f, indent=2)
```
